```python
import numpy as np
from scipy.optimize import *

import matplotlib.pylab as plt
%matplotlib inline
```

```python
def fun(x, t, y):
    return x[0] + x[1] * np.exp(x[2] * t) - y

x0 = np.array([1.0, 1.0, 0.0])
```

```python
points_x = np.array(np.linspace(-100,100,100))  #np.array(np.linspace(-100,100,100))
points_y_real = 2*points_x**2 - 34*points_x + 4 
points_y = points_y_real + np.random.normal(0,4000,len(points_y_real))
plt.scatter(points_x, points_y)
```

```python
res_lsq = least_squares(fun, x0, args=(points_x, points_y))
res_lsq
```

```python
points_y_predict = res_lsq.x[0] + res_lsq.x[1] * np.exp(res_lsq.x[2] * points_x)
plt.scatter(points_x, points_y)
plt.plot(points_x, points_y_predict)
```

```python
def fun(x, t, y):
    return x[0] + x[1] * t + x[2] * t**2 - y

x0 = np.array([1.0, 1.0, 0.0])
```

```python
points_x = np.array(np.linspace(-100,100,100))  #np.array(np.linspace(-100,100,100))
points_y_real = 2*points_x**2 - 34*points_x + 4 
points_y = points_y_real + np.random.normal(0,4000,len(points_y_real))
plt.scatter(points_x, points_y)
```

```python
res_lsq = least_squares(fun, x0, args=(points_x, points_y))
res_lsq
```

```python
points_y_predict = res_lsq.x[0] + res_lsq.x[1] * points_x + res_lsq.x[2] * points_x**2
plt.scatter(points_x, points_y)
plt.plot(points_x, points_y_predict)
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

def func(x, a, b, c):
    return a * x**2 + b * x + c

xdata = np.linspace(0, 4, 50)
y = func(xdata, 2.5, 1.3, 0.5)
ydata = y + 0.5 * np.random.normal(size=len(xdata))

popt, pcov = curve_fit(func, xdata, ydata)

print popt

func(xdata, *popt)

plt.scatter(xdata, ydata)
plt.plot(xdata, func(xdata, *popt), color='red')
plt.show()
```
