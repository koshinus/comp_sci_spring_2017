# Раздел 4. Особенности импорта в Python

```python
# Эта магия позволит вам подгружать ваши файлы каждый раз заново
%load_ext autoreload
%autoreload 1
```

```python
import import_me
import_me.my_variable
```

Почему это важно?

Иногда нужно скомпоновать большие классы и блокноты. Динамический импорт позволит вам делать это УДОБНО!
