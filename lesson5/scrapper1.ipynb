```python
import numpy as np
import pandas as pd
```

```python
from lxml import html
import requests
```

```python
from tqdm import tqdm
```

```python
page = requests.get('https://ru.wikipedia.org/wiki/Rammstein')
tree = html.fromstring(page.content)
```

```python
data_link = tree.xpath('//a[@href]/@href')
data_link
```

```python
elements = tree.xpath('//a[@href]')
len(elements), elements
```

```python
links = []
text = []
for el in tqdm(elements):
    links.append(el.xpath('@href')[0] if len(el.xpath('@href'))>0 else None)
    text.append(el.xpath('text()')[0] if len(el.xpath('text()'))>0 else None)
```

```python
series = pd.DataFrame({"Links":links, "Text":text})
series
```

```python
series[series.Text.isnull() ]
```

```python
series_filtered = series[~series.Text.isnull() & ~series.Links.isnull()]
series_filtered
```
