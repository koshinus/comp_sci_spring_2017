```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# Этот код позволит использовать русские буквы

from matplotlib import rc
font = {'family': 'Ubuntu', # Напишите здесь шрифт, который точно есть в системе
        'weight': 'normal'}
rc('font', **font)
```

```python
x = np.linspace(0,10,1000)
y = 3*x**3 + 120*np.sin(12*x)
```

```python
fig = plt.figure(figsize=(8, 6))
ax_0 = fig.add_subplot(231)
ax_1 = fig.add_subplot(235)
```

```python
fig = plt.figure(figsize=(8, 6))
ax_0 = fig.add_subplot(231)
ax_0.plot(x,y)
ax_1 = fig.add_subplot(236)
ax_1.plot(y,x)
```

```python
fig = plt.figure()
fig.suptitle(u'Важность правильного питания!', fontsize=12, fontweight='bold')
ax = fig.add_subplot(111)
ax.set_title(u'Котлетно питательная составляющая вопроса', fontsize=12, fontweight='bold')
ax.set_xlabel(u'Степень сытости')
ax.set_ylabel(u'Количество котлет')
ax.plot(x,y)
```
