```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.linspace(0,10,1000)
y = 3*x**3 + 120*np.sin(12*x)
```

```python
%matplotlib inline
plt.plot(x,y)
```

```python
# Этот код позволит использовать русские буквы

from matplotlib import rc
font = {'family': 'Ubuntu', # Напишите здесь шрифт, который точно есть в системе
        'weight': 'normal'}
rc('font', **font)
```

```python
%matplotlib inline
plt.xlabel(u'Количество котлет')
plt.ylabel(u'Степень сытости')
plt.plot(x,y)
```

```python
%matplotlib inline
plt.title(u'Котлетная составляющая вопроса')
plt.suptitle(u'Важность правильного питания!', fontsize=12, fontweight='bold')
plt.xlabel(u'Количество котлет')
plt.ylabel(u'Степень сытости')
plt.plot(x,y)
```

```python
%matplotlib inline
plt.title(u'Котлетная составляющая вопроса')
plt.suptitle(u'Важность правильного питания!', fontsize=12, fontweight='bold')
plt.text(1.5, 2000, r'$\cos(2 \pi t) \exp(-t)$', fontsize=18) # Кординаты привязаны к данным
plt.xlabel(u'Количество котлет')
plt.ylabel(u'Степень сытости')
plt.plot(x,y)
```
